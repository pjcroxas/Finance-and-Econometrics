```python

# eliminate special characters in gdp growth rate
un_stat['GDP growth rate (annual %, const. 2005 prices)'] = un_stat['GDP growth rate (annual %, const. 2005 prices)'].replace('~', '', regex=True)
```

```python
# replace "~" to empty string
un_stat["GDP growth rate (annual %, const. 2005 prices)"] = un_stat["GDP growth rate (annual %, const. 2005 prices)"].replace('~', '', regex=True)

# Convert objects to float
un_stat["GDP growth rate (annual %, const. 2005 prices)"] = un_stat["GDP growth rate (annual %, const. 2005 prices)"].astype(float)
```

```python
gdp_per_region = un_stat.groupby("Region").agg({
    "GDP: Gross domestic product (million current US$)":"mean",
    "GDP growth rate (annual %, const. 2005 prices)":"mean",
    "GDP per capita (current US$)":"mean",
    "Population in thousands (2017)":"sum"    
}).reset_index()

gdp_per_region.sort_values(by="Population in thousands (2017)",ascending=False, inplace=True)
gdp_per_region.head()
```

1. `GDP: Gross Domestic Product (million current US$)` – The total monetary value of all final goods and services produced within a country’s borders in a given year, measured in current U.S. dollars (i.e., not adjusted for inflation).

2. `GDP Growth Rate (annual %, constant 2005 prices)` – The percentage change in a country’s GDP from one year to the next, adjusted for inflation using 2005 as the base year. This measure reflects real economic growth by accounting for price changes over time.

3. `GDP per Capita (current US$)` – The GDP divided by the total population of a country, expressed in current U.S. dollars. It provides an estimate of the average economic output or income per person in a given year, without adjusting for inflation.

# Trade and Economic Balance

```python
def replace_convert(char, data, feature):
    # replace char to empty string
    data[feature] = data[feature].replace(char, '0', regex=True)
    return data    
```

```python
un_stat_trade = un_stat.copy()
trade_cols = ['International trade: Exports (million US$)',
              'International trade: Imports (million US$)',
              'International trade: Balance (million US$)']


for cols in trade_cols:
    replace_convert('...', un_stat_trade, cols)
    
for cols in trade_cols:
    replace_convert('~', un_stat_trade, cols)

for cols in trade_cols:
    un_stat_trade[cols] = un_stat_trade[cols].astype(float)
```

```python
px.box(un_stat_trade,
          x="Region",
          y="International trade: Imports (million US$)",
          # text="country",
          color="Region",
          log_y=True).update_layout(template="plotly_white",
                                   title="Imports per Region")
```

```python
px.box(un_stat_trade,
          x="Region",
          y="International trade: Exports (million US$)",
          # text="country",
          color="Region",
          log_y=True).update_layout(template="plotly_white",
                                   title="Exports per Region")
```

```python
gdp_trade = un_stat_trade.groupby('Region').agg({
    'GDP: Gross domestic product (million current US$)':'mean',
    'GDP per capita (current US$)':'mean',
    'International trade: Exports (million US$)':'mean',
    'International trade: Imports (million US$)':'mean',
    'International trade: Balance (million US$)':'mean',
    
}).reset_index()
gdp_trade.head()
```

```python
fig = px.scatter(gdp_trade,
            x="International trade: Exports (million US$)",
            y="GDP: Gross domestic product (million current US$)",
            color="Region",
            log_x=True,
            size="GDP: Gross domestic product (million current US$)",
            text="Region").update_layout(template="plotly_white")

fig.show()
```

```python
gdp_per_capita_trade_balance = gdp_trade["International trade: Balance (million US$)"].corr(gdp_trade["GDP per capita (current US$)"])
fig = px.scatter(gdp_trade,
            x="International trade: Imports (million US$)",
            y="GDP: Gross domestic product (million current US$)",
            color="Region",
            log_x=True,
            size="GDP: Gross domestic product (million current US$)",
            text="Region").update_layout(template="plotly_white",
                                        title=f"International Trade vs GDP <br><sup>corr (GDP per Capita vs Trade Balance) : {gdp_per_capita_trade_balance}</sup>")

fig.show()
```
